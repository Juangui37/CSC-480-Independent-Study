```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file with query results
results_df = pd.read_excel("QueriesCombined.xlsx")

# Define colors for each database
mongodb_color = "green"
mysql_color = "blue"

# Loop through each query and create a side by side bar chart
for i, row in results_df.iterrows():
    query = row["Query"]
    mongodb_result = row["MongoDB Result"]
    mysql_result = row["MySQL Result"]
    plt.bar(["MongoDB", "MySQL"], [mongodb_result, mysql_result], color=[mongodb_color, mysql_color])
    plt.title(f"Query {query}")
    plt.xlabel("Database")
    plt.ylabel("Execution Time (seconds)")
    plt.text(0, mongodb_result, f"{mongodb_result:.15f}", ha="center", va="bottom")
    plt.text(1, mysql_result, f"{mysql_result:.15f}", ha="center", va="bottom")
    plt.xticks([0, 1], [f"MongoDB - Query {i+1}", f"MySQL - Query {i+1}"])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file with query results
results_df = pd.read_excel("QueriesCombined.xlsx")

# Define colors for each database
mongodb_color = "green"
mysql_color = "blue"

# Loop through each query and create a side by side bar chart
for i, row in results_df.iterrows():
    query = row["Query"]
    mongodb_result = row["MongoDB Result"]
    mysql_result = row["MySQL Result"]
    plt.figure(figsize=(10, 6)) # set the figure size
    plt.bar(["MongoDB", "MySQL"], [mongodb_result, mysql_result], color=[mongodb_color, mysql_color])
    plt.title(f"Query {query}")
    plt.xlabel("Database")
    plt.ylabel("Execution Time (seconds)")
    plt.text(0, mongodb_result, f"{mongodb_result:.15f}", ha="center", va="bottom")
    plt.text(1, mysql_result, f"{mysql_result:.15f}", ha="center", va="bottom")
    plt.xticks([0, 1], [f"MongoDB - Query {i+1}", f"MySQL - Query {i+1}"])
    plt.savefig(f"query_{i+1}.png", bbox_inches='tight') # save the figure with a tight bounding box
    plt.close() # close the figure to free up memory
```
