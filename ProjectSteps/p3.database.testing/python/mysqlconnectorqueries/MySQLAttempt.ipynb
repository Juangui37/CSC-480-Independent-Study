```python
import mysql.connector
from datetime import datetime
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="admin",
  password="password",
  database="housing"
)
```

```python
mycursor = mydb.cursor()
```

```python
mycursor.execute("SHOW TABLES")
tables = mycursor.fetchall()
for table in tables:
    print(table)
```

```python
mycursor.execute("SELECT COLUMN_NAME FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_SCHEMA = 'housing' AND TABLE_NAME = 'PropertyInfo'")
columns = mycursor.fetchall()
print(columns)
```

```python
mycursor.execute("SELECT * FROM PropertyInfo")
myresult = mycursor.fetchall()
for row in myresult:
  print(row)
```

```python
# Query 1 Find all the MLSNum in PropertyInfo table with ListPrice greater than $500,000.
mycursor.execute("SELECT MlSNum, HouseStatus, Address, City, Acres, YearBuilt FROM PropertyInfo WHERE ListPrice > 500000")
result = mycursor.fetchall()
print("Results for Query 1:")
for row in result:
    print(row)
```

```python
#Query 2 Find all the distinct PropType in PropertyInfo table.
mycursor.execute("SELECT DISTINCT PropType FROM PropertyInfo")
result = mycursor.fetchall()
print("Results for Query 2:")
for row in result:
    print(row)
```

```python
# Query 3 Find the number of houses with HouseStatus of "For Sale" in PropertyInfo table.
mycursor.execute("SELECT COUNT(*) FROM PropertyInfo WHERE HouseStatus = 'ACTV'")
result = mycursor.fetchone()
print(f"Results for Query 3: {result[0]}")
```

```python
# Query 4 Find the average ClosePrice for each City in PropertyInfo table.
mycursor.execute("SELECT City, AVG(ClosePrice) FROM PropertyInfo GROUP BY City")
result = mycursor.fetchall()
print("Results for Query 4:")
for row in result:
    print(row[0], row[1])
```

```python
# Query 7 Find the total number of houses that each Realtor has in the RealtorListing table.
mycursor.execute("SELECT AgentID, COUNT(*) as TotalListings FROM RealtorListing GROUP BY AgentID")
result = mycursor.fetchall()
print("Results for Query 7:")
for row in result:
    print(row[0], row[1])
```

```python
# Query 8 Find the number of houses with HouseStatus of "For Sale" in each City in PropertyInfo table.
mycursor.execute("SELECT City, COUNT(*) as TotalForSale FROM PropertyInfo WHERE HouseStatus = 'ACTV' GROUP BY City")
result = mycursor.fetchall()
print("Results for Query 8:")
for row in result:
    print(row[0], row[1])
```

```python
# Query 9 Find the average ListPrice for each PropType in PropertyInfo table.
mycursor.execute("SELECT PropType, AVG(ListPrice) FROM PropertyInfo GROUP BY PropType")
result = mycursor.fetchall()
print("Results for Query 9:")
for row in result:
    print(row[0], row[1])
```

```python
# Query 10 Find the top 10 most common ListingAgents in the Realtor table.
mycursor.execute("SELECT ListingAgent, COUNT(*) as TotalListings FROM Realtor GROUP BY ListingAgent ORDER BY TotalListings DESC LIMIT 10")
result = mycursor.fetchall()
print("Results for Query 10:")
for row in result:
    print(row[0], row[1])
```

```python
# Query 11 Find the average ClosePrice for each year a house was built in PropertyInfo table.
mycursor.execute("SELECT YearBuilt, AVG(ClosePrice) FROM PropertyInfo GROUP BY YearBuilt ORDER BY YearBuilt")
result = mycursor.fetchall()
print("Results for Query 11:")
for row in result:
    print(row[0], row[1])
```
