```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from random import randint, choice

# define the possible values for Status
status_values = ['ACTV', 'UC-CTS', 'UC', 'CLSD', 'NEW', 'REDC', 'RAIS']

# Define possible values for City attribute
city_list = ['Avon', 'Berlin', 'Bloomfield', 'Bolton', 'Cromwell', 'East Hartford',
             'East Windsor', 'Farmington', 'Glastonbury', 'Hartford', 'Manchester',
             'Marlborough', 'Middletown', 'New Britain', 'Newington', 'Plainville',
             'Portland', 'Rocky Hill', 'Simsbury', 'South Windsor', 'Vernon',
             'West Hartford', 'Wethersfield', 'Windsor', 'Windsor Locks']


# read in the original file
df = pd.read_excel('MLS_Final.xlsx')

# Define the range of possible list and close prices
list_price_range = (90000, 500000)
close_price_range = (95000, 515000)


# create a function to generate random data
def generate_data():
    MLSNum = randint(170000000, 179999999)
    Status = random.choice(status_values)
    Status_Change_Timestamp = datetime.now() - timedelta(days=randint(1, 365))
    PropType = choice(['SF'])
    List_Price = randint(100000, 1000000)
    Close_Price = List_Price - randint(1000, 10000)
    Address = 'Mock Address ' + str(randint(1, 10000))
    City = random.choice(city_list)
    Acres = round(np.random.normal(1, 0.5), 2)
    Sq_Ft_Total = randint(500, 5000)
    Sq_Ft_Est_Heated_Above_Grade = randint(500, 5000)
    Style_or_Rent_Type = choice(['Cape Cod', 'Colonial', 'Ranch', 'Raised Ranch'])
    Rooms_Total = randint(5, 10)
    Beds_Total = randint(1, 5)
    Bathrooms = round(np.random.normal(2, 0.5), 1)
    Garage_Park = choice(['1 Car/Attached Garage', '2 Car/Attached Garage', 'Car/Unpaved', 'Car/Paved'])
    Year_Built = randint(1900, 2023)
    Days_on_Market = randint(0, 365)
    AgentID = randint(1000, 9999)
    Listing_Agent = 'Mock Agent ' + str(randint(1, 100))
    Phone_Number = randint(1000000000, 9999999999)
    Email = Listing_Agent.lower().replace(' ', '') + '@example.com'
    return [MLSNum, Status, Status_Change_Timestamp, PropType, List_Price, Close_Price,
            Address, City, Acres, Sq_Ft_Total, Sq_Ft_Est_Heated_Above_Grade, Style_or_Rent_Type,
            Rooms_Total, Beds_Total, Bathrooms, Garage_Park, Year_Built, Days_on_Market,
            AgentID, Listing_Agent, Phone_Number, Email]

# generate 3000 rows of data
new_data = []
for i in range(3000):
    new_data.append(generate_data())

# add the new data to the original dataframe
df_new = pd.concat([df, pd.DataFrame(new_data, columns=df.columns)], ignore_index=True)

# save the updated dataframe to the same file
df_new.to_excel('MLS_Final1.xlsx', index=False)
```
